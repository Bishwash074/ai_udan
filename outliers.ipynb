```python
import pandas as pd
```

```python
dd=pd.read_csv("dataset/bad_daraz_dataset.csv")
dd
```

```python
dd["price"]=pd.to_numeric(dd["price"],errors="coerce")

# Quartile(q3),q1-->IQR
```

```python
mean=dd["price"].mean()
std=dd["price"].std()
dd["z_score"]=(dd["price"]-mean)/std #z-score>2 outliers
outlier=dd[dd["z_score"].abs()>2]
outlier
```

```python
Q1=dd["price"].quantile(0.25)
Q3=dd["price"].quantile(0.75)
median=dd["price"].quantile(0.5)
IQR=Q3-Q1
lower_bound=Q1-(1.5*IQR)
upper_bound=Q3+(1.5*IQR)
upper_bound
```

```python
outlier=dd[(dd["price"]<lower_bound)| (dd["price"]>upper_bound)]
outlier
```

```python
len(outlier)
```

```python
dd.loc[(dd["price"]<lower_bound)|(dd["price"]>upper_bound)]=median
dd
```

```python
# IQR fix
# winsorization method
dd["price"].clip(lower_bound,upper_bound)
```
